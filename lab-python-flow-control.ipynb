```python
# Paso 1
products = ["t-shirt", "mug", "hat", "book", "keychain"]

# Paso 2
inventory = {}

# Paso 3


cantidad_tshirt = int(input("¿Cuánt@s t-shirt hay en inventario? "))
cantidad_mug = int(input("¿Cuánt@s mug hay en inventario? "))
cantidad_hat = int(input("¿Cuánt@s hat hay en inventario? "))
cantidad_book = int(input("¿Cuánt@s book hay en inventario? "))
cantidad_keychain = int(input("¿Cuánt@s keychain hay en inventario? "))

inventory["t-shirt"] = cantidad_tshirt
inventory["mug"] = cantidad_mug
inventory["hat"] = cantidad_hat
inventory["book"] = cantidad_book
inventory["keychain"] = cantidad_keychain

# Paso 4
customer_orders = set()

# Paso 5
producto1 = input("Introduce el primer producto que desea pedir: ")
producto2 = input("Introduce el segundo producto que desea pedir: ")
producto3 = input("Introduce el tercer producto que desea pedir: ")

customer_orders.add(producto1)
customer_orders.add(producto2)
customer_orders.add(producto3)

# Paso 6
print("Productos pedidos:")
for producto in customer_orders:
    print(producto)

# Paso 7: Calcular las estadísticas del pedido
total_products_ordered = len(customer_orders)          # Total de productos pedidos
total_available_products = len(products)               # Total de productos disponibles
percentage_ordered = (total_products_ordered / total_available_products) * 100
order_status = (total_products_ordered, percentage_ordered)  # Tupla con las estadísticas

# Paso 8
print("Order Statistics:")
print(f"Total Products Ordered: {order_status[0]}")
print(f"Percentage of Products Ordered: {order_status[1]:.2f}%")

# Paso 9
if "t-shirt" in customer_orders:
    inventory["t-shirt"] -= 1
if "mug" in customer_orders:
    inventory["mug"] -= 1
if "hat" in customer_orders:
    inventory["hat"] -= 1
if "book" in customer_orders:
    inventory["book"] -= 1
if "keychain" in customer_orders:
    inventory["keychain"] -= 1

# Paso 10
print("Inventario actualizado:")
print("t-shirt:", inventory["t-shirt"])
print("mug:", inventory["mug"])
print("hat:", inventory["hat"])
print("book:", inventory["book"])
print("keychain:", inventory["keychain"])
```

```python
# Paso 1
products = ["t-shirt", "mug", "hat", "book", "keychain"]

# Paso 2
inventory = {}

for product in products:
    cantidad = int(input(f"¿Cuánt@s {product} hay en inventario? "))
    inventory[product] = cantidad


customer_orders = set()

while True:
    producto = input("Introduce el producto que desea pedir: ")
    customer_orders.add(producto)

    otro = input("¿Quieres añadir otro producto? (yes/no): ").lower()

    if otro == "no":
        break

#Paso 3

for product in customer_orders:
    inventory[product] -= 1
    
```
